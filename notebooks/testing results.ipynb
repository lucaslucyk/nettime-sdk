```python
loa = []

employeeFields = ["id", "code", "name"]
results = [
    "Saldo_Franco_Compensatorio",
    "EXTRAS_50",
    "Hs_Extras_50_Aprob",
    "Hs_EXTRAS_50_Aprob_SAB",
    "EXTRAS_100",
    "Hs_Extras_100_Aprob",
    "HS_EXTRAS_100_Aprob_SAB"
]

for e in loe:
    structure = {
        "employee": {},
        "totals": dict(zip(results, e.get("values"))),
    }
    
    # update person fields
    for i in range(len(employeeFields)):
        structure.get("employee").update({
            employeeFields[i]: e.get("dimKey")[i]
        })
    
    # if results per date -or month-
    if 'children' in e.keys():
        structure["frame"] = pd.DataFrame(
            [child.get("values") for child in e.get("children")],
            columns=results,
            index=[child.get("dimKey")[0] for child in e.get("children")]
        )#.T.to_dict()
        
    loa.append(structure)
```

```python
loa[0]["frame"].to_csv('out_data.csv', sep=";")#, mode="a")
```
